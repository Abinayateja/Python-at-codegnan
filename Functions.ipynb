```python
# Types of functions based on parameters and return value
# 4. With parameters and with return 
# funtion definition block
def addition(x,y):
    c = x + y
    return c
# funtion calling block
a,b= 10,20
result = addition(a,b)
print(result)
```

```python
# 3. With parameters and without return 
# funtion definition block
# By default function returns None value, when function not returning any value.
def addition(x:int,y:int)->None:
    c = x + y
    print(f"This is from function definition blok: {c}")
# funtion calling block
a,b= 10,20
result = addition(a,b)
print(f"This is from function calling block: {result}")
```

```python
# 2. Without parameters and with return 
# We should not take input in the funtion definition block.
# funtion definition blok
def addition()->int:
    x = 10
    y = 20
    z = x + y
    return z
# funtion calling block
result = addition()
print(f"This is from function calling block: {result}")
```

```python
# 1. Without parameters and without return 
# funtion definition blok
def addition()->None:
    print("Function definition block is Executed")
# funtion calling block
result = addition()
print("This is from function calling block is Executed:",result)
```

```python
def even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
n = int(input())
res = even_odd(n)
print(f"{n} is {res}.")
```

```python
def prime_number(n):
    if n < 2:
        return "Not Prime"
    for i in range(2,n//2 + 1):
        if n % i == 0:
            return "Not Prime"
    return "Prime"
    
n = int(input())
res = prime_number(n)
print(f"{n} is {res}")
```
