```python
import re

```

```python
# match(pattern,string): It cheks whether the string starts with pattern or not.
string = "The sky is blue and ocean is blue"
res = re.match("The",string)
print(res)
print(res.span()) # It returns the matched pattern positions as tuple
print(res.group()) # It returns the matched result
```

```python
# match
try:
    string = "The sky is blue and ocean is blue"
    res = re.match("sky",string)
    print(res)
    print(res.span()) # It returns the matched pattern positions as tuple
    print(res.group()) # It returns the matched result
except:
    print(f"There is no Matched Pattern")
```

```python
# search
try:
    string = "The sky is blue and ocean is blue"
    res = re.search("sky",string)
    print(res)
    print(res.span()) # It returns the matched pattern positions as tuple
    print(res.group()) # It returns the matched result
except:
    print(f"There is no Matched Pattern")
```

```python
# Email Validation
import re
try:
    email = input()
    res = re.fullmatch(r"^[a-z0-9][a-z0-9.]+\@[a-z0-9]+\.[a-z]{2,3}$",email)
    if res:
        print(f"{email} is valid")
    else:
        print(f"{email} is invalid")
except:
    print("There is no matched pattern")
```

```python
# Date validation
try:
    date = input()
    res = re.fullmatch(r"^(0?[1-9]|[12][0-9]|3[01])[/-](0?[1-9]|1[0-2])[/-](\d{4})",date)
    if res:
        print(f"{date} is valid")
    else:
        print(f"{date} is invalid")
except:
    print("There is no matched pattern")
```
