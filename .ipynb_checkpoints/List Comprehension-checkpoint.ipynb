### List Comprehension
- This is the one way of creating list with an expression
#### Syntax
- [expression for `iterator` in iterable if condition]

```python
# First n natural numbers
n =int(input())
res = [i for i in range(1,n+1)]
print(res)
```

```python
# First n even numbers
n = int(input())
res = [i*2 for i in range(n)]
print(res)
```

```python
# First n cubic numbers
n = int(input())
res = [i**3 for i in range(1,n+1)]
print(res)
```

```python
# Find the odd numbers between n1 and n2
n1,n2 = map(int,input().split())
res = [i for i in range(n1,n2+1) if i % 2 == 1]
print(res)
```

```python
import math
arr = list(map(int,input().split()))
rst = [x** 2 if x % 2 == 0 else round(math.sqrt(x),2) for x in arr]
print(rst)
```

```python
# Two dice combination values
res = [(x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
# Two dice combination values if the sum of two values divisible by 3
res = [(x,y) for x in range(1,7) for y in range(1,7) if (x + y) % 3 == 0]
print(res)
```

```python
# Two dice combination values if any one value is 4
res = [(x,y) for x in range(1,7) for y in range(1,7) if x == 4 or y ==4]
print(res)
```

```python
# count the number of vowels exists in string.
string = input()
vowels = "aeiouAEIOU"
res = [x for x in string if x in vowels]
print(len(res))
```

```python
n = int(input())
print({x for x in range(1, n+1)})
```

```python
n = int(input())
print({x:x**2 for x in range(1, n+1)})
```

```python
n = int(input())
print(tuple(x for x in range(1, n+1)))
```

```python
# divmod() : returns quotient and remainder
print(divmod(4,2))
```

```python
# Check wheather a person is eligible for vote
age = int(input())
print("Eligible" if age > 18 else "Not Eligible")
```

```python
# assert keyword : This keyword is used for testing
# Check weather a person is eligible for vote using assert
age = int(input())
assert(age > 18), "Age must be above 18"
print("Eligible")
```

```python
n = int(input())
res = ""
rev_res = ""
while(n > 0):
    rem = n % 2
    res = str(rem) + res
    n = n // 2
    rev_res = str(rem) + rev_res
print("Palindrome" if res == rev_res else "Not Palindrome")
```

```python
w = input()
res = ""
for i in w:
    res =  i + res
print(res)
```
