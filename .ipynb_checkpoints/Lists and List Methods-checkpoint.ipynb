# Lists

```python
# Define list
lst = [1,2,3,4.5,"Hi","Hello",1,3,4]
print(lst)
```

```python
# list allows indexing and Slicing
# Accessing list items
lst = [1,2,4.5,"Hi","Hello",1,3,4]
print(lst[0])
print(lst[1:4])
print(len(lst)-1) # to access last element from the list
print(lst[::-1])
print(lst[:3:2])
print(lst[-3::2])
print(lst[-3:-2])
print(lst[::])
print(lst[:])
print(lst[2:-4])
print(lst[-2::])
print(lst[:-2:-1])
```

```python
# Empty list Creation
l1 = []
l2 = list()
print(type(l1),type(l2))
print(l1,l2)
```

```python
# List Operations
# 1. List Conatination (using '+' operator)
# 2. List Repetition (using '*' operator)
l1 = [1,2,3]
l2 = [3,4,5,6]
print(l1 + l2)
print(l1 * 3)
print([0] * 10)
```

### List Methods
- Item adding Methods
  - append(item): It adds item at end of list.
  - extend(iterator): It adds the all iterator elements at end of list.
  - insert(pos,item): It adds item at specified position.

```python
lst = [10,20,30,40]
lst.append('Hello')
print(lst)
char = ['a','b','c']
lst.extend(char) # This line won't return any value!!(append,extend,insert won't return any value.)
print(lst)
lst.append([100,200])
print(lst)
print(type(lst[-1]))
lst.insert(0,1000)
print(lst)
```

- Item Removing Methods
    - remove(item): It Removes the Item from the list if it is exists in list.
      othervise `ValueError`
    - pop(pos): It removes and return the specified position item if position is passed, otherwise
      it remove and returns the last item from list.
    - clear(): It removes all elements from list.

```python
lst = [10,20,30,40,50]
lst.remove(20)
print(lst)
#lst.remove(100)
val = lst.pop()
print(val)
print(lst)
val2 = lst.pop(0)
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
# index and count()
# index(item): If item is present in list it returns first ocurane of the index otherwise "Value Error"
# count(item): If return the count of Total occurances of item otherwise retrns 0.
lst = [10,20,30,20,40,50]
ind = lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# Sorting items
# sort(reverse = False): By default it sorts the original list in Ascending order.
#                          if reverse is True it sorts in Decending order.
lst = [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
# Builtin Functions len(),min(), max(), sum() works for non-premitive data types
lst = [1,2,3,4,5]
print(len(lst),min(lst),max(lst),sum(lst))
print(sum(lst) / len(lst)) # Average of the list
lst2 = ['a','a','a','a','b','c','z']
print(len(lst2))
print(lst2.count('a'))
print(max(lst2))
print(min(lst2))
```

# Note :
- methods that return value:
  - pop()
  - index()
  - find()
  - sort()
  - count()
 
- methods that don't return value:
  - clear()
  - append()
  - insert()
  - extend()
  - remove()
