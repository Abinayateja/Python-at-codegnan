# Type Conversions
- Implicit Type Conversion
- Explicit Type Conversion(`Type Casting`)

```python
num = 12
print(num/2)
```

### Explicit Type Conversion(Type Casting)

```python
# integer to float, boolean, string
num = 10
print(float(10))
print(bool(num))
print(str(num))
print(complex(num))
```

```python
# float to int, boolean, string
num = 12.34
print(int(num))
print(bool(num))
print(str(num))
```

```python
# bool to int, float, string
bool1 = True
print(int(bool1))
print(float(bool1))
print(str(bool1))
```

```python
# string to int,float,list,tuple,set
string = "123.456"
print(int(float(string)))
print(float(string))
print(bool(string))
print(list(string))
print(tuple(string))
print(set(string))
```

```python
# list to string,tuple,set
lista = [1,2,3,4,5,6]
print(str(lista))
print(tuple(lista))
print(set(lista))
```

```python
# set to string,list,tuple
seta = {1,2,3,5,6,5}
print(str(seta))
print(list(seta))
print(tuple(seta))
```

```python
# tuple to string,list,set
tuplea = (1,2,3,4,5,5,6)
print(tuplea)
print(str(tuplea))
print(list(tuplea))
print(set(tuplea))
```

```python
# dictionary to string,tuple,set,list
dicta = {'a' : 1,'b' : 2, 'c' : 'hi'}
print(str(dicta))
print(tuple(dicta))
print(set(dicta))
print(list(dicta))
```
