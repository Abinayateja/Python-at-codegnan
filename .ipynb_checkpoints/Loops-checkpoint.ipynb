```python
# Even or odd without using loops
n1,n2,n3 = map(int,input().split())
if n1 % 2 == 0:
    print("Even")
else:
    print("Odd")
if n2 % 2 == 0:
    print("Even")
else:
    print("Odd")
if n3 % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Even -> square , odd -> 
lst = list(map(int,input().split()))
for num in lst:
    if num % 2 == 0:
        print(num ** 2)
    else:
        print(num + 1)
```

```python
# Fizz(3) Buzz(5) FizzBuzz(3 and 5)
lst = list(map(int,input().split()))
for num in lst:
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

```python
# Count the vowels and consonants in a string.
word = input()
vowel = "aeiouAEIOU"
v,c = 0,0
for ch in word:
    if ch in vowel:
        v += 1
    else:
        c += 1
print(c)
print(v)
        
```

```python
# Count the vowels and consonants in a string excluding digits and special characters.
word = input()
vowel = "aeiouAEIOU"
v,c = 0,0
for ch in word:
    if ch.isalpha():
        if ch in vowel:
            v += 1
        else:
            c += 1
print(c)
print(v)
        
```

```python
# Sum of N natural numbers using for loop
n = int(input())
for i in range(1,n+1):
    print(i,end = " ")
```

```python
# Sum of N natural numbers using while loop.
n = int(input())
i = 1
while i <= n:
    print(i,end = " ")
    i += 1
```

```python
# Sum of N natural numbers using while loop.
n = int(input())
i = 1
while i <= n:
    print(n,end = " ")
    n -= 1
```

```python
total_budget = 150
pen_cost = 8
eraser_cost = 5
sharpner_cost = 4
pens_brought = 10
erasers_brought = 10
pen_budget = pen_cost * pens_brought
eraser_budget = eraser_cost * erasers_brought
sharpners_brought = (total_budget - (pen_budget + eraser_budget)) // 4
sharpners_budget = sharpners_brought * sharpner_cost
total_items = pens_brought + sharpners_brought + erasers_brought
print(f"Average price : {total_budget / total_items}")
```

```python
lst = list(map(int,input().split()))
target = int(input())
for i in range(len(lst)):
    if lst[i] == target:
        print(i)
        break
    else:
        print(-1)
        break
```

```python
n = int(input())
fac = 2
factors = [1,n]
for i in range(2,(n//2) + 1):
    if n % i == 0:
        factors += [i]
        fac += 1
        break
if fac == 2:
    print("Prime")
    print(factors)
else:
    print("Not Prime")
```

```python
n = int(input())
for i in range(2,(n//2) + 1):
    if n % i == 0:
        print("Not Prime")
        break
else:
    print("Prime")
```

```python
# Nested Loops
n = int(input())
for i in range(n):
    spaces = n - i
    stars = i
    for j in range(n):
        print(" " * spaces, end=" ")
        print("*" * stars,end = " ")
    print()
```
