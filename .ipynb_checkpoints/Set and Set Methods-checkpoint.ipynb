# Set Characteristics

```python
# Set unordered collection.
# Set return unique elements.
s = {1,2,3,2,3,'hi','a',2.02,(10,20,30)}
print(s)
# print(s[0]) # set does not supports indexing.
```

```python
# s = {1,2,3,2,3,'hi','a',2.02,(10,20,30),[1,2,3]} # list can't be used as set element.
print(s)
```

```python
# Empty Set
s1 = {}
s2 = set()
print(type(s1),type(s2))
```

### Set Methods

```python
# Elements adding method
# add(ele): It adds the element to set if element not in set, otherwise nothing
s = {1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("Code")
print(s)
```

```python
# Element removing methods
# remove(item): it removes the item if it exists in set otherwise 'KeyError'.
# discard(item): It removes the item if it exists in the set otherwise nothing.
s = {1,2,3,4,5}
s.remove(5)
print(s)
# s.remove(100) # KeyError
s.discard(3)
print(s)
s.discard(100)
```

```python
# sets Operations
# union(set): It returns the all unique elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
union = s1.union(s2,s3)
print(union)
res = s1 | s2 | s3 # we can also use the set symbol
print(res)
```

```python
# intersection(set): Returns all common elements from all sets
s1 = {1, 3, 5, 7}
s2 = {4, 5, 6, 1}
s3 = {1, 2, 5, 9}
intersection = s1.intersection(s2, s3)
print(intersection)
res = s1 & s2 & s3  # We can also use the & operator
print(res)
```

```python
# difference(set): Returns values that are not present in remaining sets.
s1 = {1, 3, 5, 7}
s2 = {4, 5, 6, 1}
s3 = {1, 2, 5, 9}
difference = s1.difference(s2, s3)
print(difference)
res = s1 - s2 - s3  # We can also use the - operator
print(res)
```

```python
# symmetric_difference(set): Returns values non-common elements from all sets.
s1 = {1, 3, 5, 7}
s2 = {4, 5, 6, 1}
symmetric_difference = s1.symmetric_difference(s2)
print(symmetric_difference)
res = s1 ^ s2  # We can also use the - operator
print(res)
```
