```python
# Check whether number is divisible by 3 or not.
num = int(input())
if num % 3 == 0:
    print(True)
else:
    print(False)
```

```python
# Check whether the number is divisible by bothe 3 and 5
num = int(input())
if num % 3 == 0 and num % 5 == 0:
    print(True)
else:
    print(False)
```

```python
# Check whether the number is even or not.
num = int(input())
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Check whether the number is divisible by either 3 or 5
num = int(input())
if num % 3 == 0 or num % 5 == 0:
    print(True)
else:
    print(False)
```

```python
# Generate total bill if total bill more than 500 then give 10% discount otherwise 0% disount.
bill = int(input())
discount = 0
if bill > 500:
    discount += bill * 0.1
    print(bill - discount)
else:
    print(bill)
```

```python
# Check weather the product stock is available or not (read product name and stock)
product = input()
stock = int(input())
if stock > 0:
    print("Stock Available")
else:
    print("Stock Not Available")
```

```python
items = input().split()
dicta = dict()
for i in items:
    dicta[i] = int(input())
required_items = input().split()
total_price = 0
for i in required_items:
    if i in required_items:
        total_price += dicta[i]
if total_price > 0:
    print(total_price)
else:
    print("Item not Availale")
```
