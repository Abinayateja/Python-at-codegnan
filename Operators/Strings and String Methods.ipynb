## String Methods
- Case Conversion Methods
  - upper(): It Converts entire string into Upper Case.
  - lower(): It Converts entire string into Lower Case.
  - title(): It Converts the first letter of every word into Upper Case and remaining into Lower Case.
  - swapcase(): It Converts the Upper Case letters into Lower Case Letters and Vice Versa.
  - capitalize(): It Converts First character into Upper Case and remaining into Lower Case.
#### Note : 
When ever we use any of string method it will create a new string object.

```python
string = "The Sky is blue and Ocean is BLUE"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- Remove white space or pattern at ends of String
  - strip(): By default it removes the white spaces from both the ends of the string.
  - lstrip(): It removes the white space at the left side of the string.
  - rstrip(): It removes the white space at the right side of the string.

```python
s = "             Codegnan        "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s = "@@@@@@@@@Cod@@egnan@@@@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
# split: It splits the entire string into list of substrings based on the specified seperator.
# by default it splits at space.
string = "The Sky is blue and Ocean is Blue"
print(string.split())
print(string.split("is"))
print(string.split('e'))
```

```python
# 'sep'.join(iterable): It joins the iterable elements with seperator.
words = ['The', 'Sky', 'is', 'blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
```

### Substring Index finding
- index(substring): if substring exists in the string, it returns the first occurance
- of substring position, otherwise `ValueError`.
- find(substring): if substring exists in the string, it returns the first occurance
- of substring position, otherwise `-1`.

```python
string = "The Sky is blue and Ocean is Blue"
print(string.index('is'))
print(string.index('z'))
```

```python
print(string.find('is'))
print(string.find('z'))
print(string.index('e',10)) # it starts checking from 10th position.
print(string.index('e',-5))
```

```python
# count(substring): It returns the total non-overlapping occurences of substring.
s = "ABCBBbbaabb"
print(s.count('B'))
print(s.lower().count('b'))
print(s.count('s'))
```

```python
# replace(old_substring,new_substring): It replaces the old substring with the new_substring within a string.
string = "The Sky is blue and Ocean is Blue" # Red Sea
print(string.replace("Blue","Red"))
print(string.replace('e','E'))
print(string.replace('e','E',2))#It replaces the first 2 occurances.
```

### String Checking Methods
- String Checking methods returns the Boolean value as result .
- isupper(): It return True, If entire string is in upper case.
- islower(): It return True, If entire string is in lower case.
- isalpha(): It return True, If entire string contains alphabets.
- isdigit(): It return True, If entire string contains digits.
- isalnum(): It return True, If entire string contains either alphabets or digits.
- startswith(substring): It return True, If string starts with the specified substring.
- endswith(substring): It return True, If String ends with the speified substring.


```python
s = "CODE"
print(s.isupper())
print(s.islower())
print("code".islower())
print(s.isalpha())
print("123".isdigit())
print("Code123".isalnum())
print("123".isalnum())
print(s.startswith("C"))
print(s.endswith("E"))
```
