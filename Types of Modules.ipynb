# Types of Modules


### 1. User Defined Modules
    The modules that are provided by the user by defining in the program.
### 2. Built in Modules
    The module that are provided by the python package.

    Examples : 
    1. Math - To perform advanced Mathematical Operations
    2. random - To generate/access Random values in a range of values
    3. os - Operating System (Creation of file or directory)
    4. sys - System (Related to the Interpreter)
    5. pandas - Data Manipulation
    6. numpy - Numerical Python operations
    7. matplotlib,seaborn - Data Visualization
    8. scikit-learn - For Machine Learning

```python
import math
print(math.ceil(4.25))   # It returns the upper bound integer value.
print(math.floor(4.25))  # It returns the lower bound integer value.
print(math.factorial(5)) # It returns the factorial of a number.
print(math.gcd(10,6,4))  # It returns the Greatest Common Devisor for the given numbers.
print(math.lcm(10,6,4))  # It returns the Least Common Multiple for the given numbers.
print(math.pow(10,3))    # pow(a,b) : It returns the Foat type of a raise to the power of b.
print(math.sqrt(9))      # It returns the square root value of a number.
print(math.log10(4))     # To find the Log10 values
```

```python
import random
print(random.random()) # Everytime it returns a random value from 0 to 1
print(random.randint(1,6)) # random(a,b) : It returns the random value between a,b
print(random.randint(1000,9999)) # 4 digit OTP generation
print(random.choice(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'])) # This function returns random value from sequence.
print(random.choices(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'],k = 3)) # This function returns random value from sequence.
arr = ['A',1,2,3,4,5,6,7,8,9,'J','Q','K']
random.shuffle(arr)
print(arr)
```

```python
# Snake and Ladder Game Logic.
import random
rolls = 0
count = 0
snakes = {24: 4,44 : 22,87:17,97:87, 98:78}
laders = {}
while(count < 100):
    input("To throw a die, please click enter button")
    rolls += 1
    A = random.randint(1,6)
    if count + A <= 100:
        count += A
    if count in snakes:
        count = snakes[count]
    if count in laders:
        count = snakes[laders]
    print(f"Roll : {rolls} \nValue : {A} \nCount : {count}")
```
