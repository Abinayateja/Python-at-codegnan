# Input Formating

```python
# reading integer from user
num = int(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# reading float from user
num = float(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# complex reading from user
val = complex(int(input()),int(input()))
print(type(val))
print(val)
```

```python
# boolean reading
val = bool(input())
print(type(val))
print(val)
```

```python
# boolean reading as integer
val = bool(int(input()))
print(type(val))
print(val)
```

```python
# reading string value from user
string = input()
print(type(string))
print(string)
```

```python
# reading list from user
lsta = list(map(int,input().split()))
print(map(int,['1']))
print(list(map(int,['1'])))
print(type(lsta))
print(lsta)
```

```python
# Reading heterogeneous values of list from user
lsta = list(map(lambda x: int(x) if x.isdigit() else x, input().split()))
print(lsta)
```

```python
# reading tuple from user
tuplea = tuple(map(int,input().split()))
print(type(tuplea))
print(tuplea)
```

```python
# reading set from user
seta = set(map(int,input().split()))
print(type(seta))
print(seta)
```

```python
# reading dictionary from user
dica = dict(map(lambda ,input().split()))
print(type(dica))
print(dica)
```

```python
# eval(): This funtion evaluates the string type expression
print("3 + 5")
print(eval("3 + 5"))
```

```python
val = eval(input())
print(type(val))
print(val)
```

```python
name =  "Abinayateja"
age = 21
print("My name is", name, "and I am", age, "years old")
print("My name is " + name + " and I am "+ str(age) + " years old")
print("My name is {} and I am {} years old".format(name,age)) # Formatted String.
print(f"My name is {name} and I am {age} years old") # Formatted String.
```

```python
lst = ['a','b','c']
dicta = {}
for x in lst:
    dicta[x] = input()
print(dicta)
```

```python
# Access Speifiers
# string - %s
# integer - %d
# float - %f
name =  "Abinayateja"
age = 21
print("My name is %s and I am %d years old"%(name,age))
# print("My name is %s and I am %d years old"%(age,name)) # Raise Error
```
