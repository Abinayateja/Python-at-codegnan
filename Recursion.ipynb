# Recursion

### Definition :
A Function calls itself either directly or indirectly.
#### Types of Recursion
- Direct Recursion
- Indirect Recursion

`1.` Direct Recursion: A function calls itself until it reaches to base case.
#### Base Condition:
- The base condition is used to stop the infinite reursive call.
#### Recursive Call:
- The function call recursively until it reaches the base condition.

```python
def fun(x):
    if x == 0:
        return 0
    print(x)
    fun(x - 1)
fun(3)
```
