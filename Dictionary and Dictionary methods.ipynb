# Dictionary Characteristics

```python
# Dictionary ordered Collection
details = {"Name": "Abi"}
```

```python
# Dictionary values acess by using keys
# updating dictionary values
# adding new items to dictionary
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print(details['course'])
# print(details['location'])
details['age'] = details['age'] + 1
print(details['age'])
details['location'] = "Hyderabad"
print(details)
```

```python
details = {"age" : 21,"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print(details['age'])
print(details)
```

```python
# List can't be used as dictionary key
details = {"name": "Abi", "course": "Python", "batch": 57, [1,2,3] : 1}
print(details)
# dictionary keys are unique
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print(details[0])
```

## Dictionary Methods
- `get(key,default_value:Optional)` It returns the value if key is exists in dictionary, otherwise it returns default value.

```python
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print(details.get('marks',"Marks not available"))
```

- `update(new_dictionary)`: This method updates the value if new_dictionary keys already exists in the Dictionary, otherwise it adds new items to dictionary 

```python
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
details.update({'batch': 58, 'course': "Java", "location": "Hyderabad"})
print(details)
```

- `keys()`: This method returns all the list of dictionary keys.
- `values()`: This method returns the list of dictionary values.
- `items()`: This method return the list of Tuple of key-value pairs.

```python
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print(details.keys())
print(details.values())
print(details.items())
```

- popitem(): It returns and removes the last item as tuple.
- pop(key): It removes the specified key item and return the specified key's value, otherwise `KeyError`

```python
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print(details.popitem())
print(details)
print(details.pop('batch'))
print(details)
```

```python
# Membership
# It checks with dictionary keys
details = {"name": "Abi", "course": "Python", "batch": 57, "age": 20}
print('name' in details)
print('Abi' in details)
```
