```python
# Generate total bill if total bill more than 500 then give 10% discount otherwise 0% disount.
bill = int(input())
discount = 0
total_bill = 0
if bill > 500:
    discount += bill * 0.1
    total_bill = bill - discount
    print(total_bill)
else:
    print(total_bill)
```

```python
items = input().split()
dicta = dict()
for i in items:
    dicta[i] = int(input())
required_items = input().split()
total_price = 0
quantities = list(map(int,input().split()))
for i,q in zip(required_items,quantities): # zip() will pair the two alternative iterables values.
    if i in dicta:
        total_price += dicta[i] * q
if total_price > 0:
    print(total_price)
else:
    print("Item not Availale")
```

```python
# Check whether the number is +ve Odd, -ve Odd, +ve Even, -ve Even
# Using "Nested if"
num = int(input())
if num >= 0:
    if(num % 2 == 0):
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    if(num % 2 == 0):
        print("Negative Even")
    else:
        print("Negative Odd")
```

```python
# Check whether the number is +ve Odd, -ve Odd, +ve Even, -ve Even
# Using "elif"
num = int(input())
if num >=0 and num % 2 == 0:
    print("Positive Even")
elif num < 0 and num % 2 == 0:
    print("Negative Even")
elif num >= 0 and num % 2 != 0:
    print("Positive Odd")
else:
    print("Negative Odd")
```

```python
# Check whether the number is +ve Odd, -ve Odd, +ve Even, -ve Even
# Using "elif(optimised)"
num = int(input())
if num >= 0 and num % 2 == 0:
    print("Positive Even")
elif num >=0:
    print("Positive Odd")
elif num % 2 == 0:
    print("Negative Even")
else:
    print("Negative Odd")
```
