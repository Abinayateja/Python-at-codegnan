```python
s = "Codegnan"
print(s[2])
print(s[-1])
```

```python
# slicing
# str_obj[start:en:step]
s = "Codegnan"
print(s[:1] + "Z" + s[-6::1])
print(s[2:7])
print(s[-4:])
print(s[5:2])
print(s[3:100])
print(s[::-1])
print(s[-3::-1])
print(s[-4::1])
```

```python
# Empty string representation
# type() function returns the type of the object
s1 = ""
s2 = str()
print(type(s1))
print(type(s2))
```

```python
# string reading from user
# input() : This function used to read input dynamically from user and this function by default read it as string data type
s = input()
print(type(s))
print(s)
```

```python
# string Operations
# 1. String Concatination (using + operator)
# 2. String Repeatition (using * operator)
s1 = "Code"
s2 = "Gnan"
s3 = s1 + s2
print(s3)
print(s1 * 3)
```
