```python
s1 = input()
s2 = input()
s = ""
for i in range(len(s1)):
    diff = abs(ord(s1[i]) - ord(s2[i]))
    if diff == 0:
        s += s1[i]
    else:
        s += chr(64 + diff)
print(s)
```

```python
s1 = input()
s2 = input()
b = ""
s = ""
for i in range(len(s1)):
    if ord(s1[i]) > ord(s2[i]):
        b += s1[i]
        s += s2[i]
    elif ord(s2[i]) > ord(s1[i]):
        b +=  s2[i]
        s += s1[i]
print(f"big: {b}")
print(f"small: {s}")
```

```python
# map(fun, iterable): It applies the functionaliy on eah element in iterable
# filter(fun, iterable): It filters the iterables, it returns the values which is not None
# find the square of list of integers
def square(x):
    return x ** 2
lst = [1,2,3,4,5]
print(list(map(square,lst)))
```

```python
# sum of the list of tuples (We can pass multiple values)
def even(t):
    x, y = t
    return x + y
arr = [(1,2),(3,4),(5,6)]
print(list(map(even,arr)))
```

```python
# filter the evnen numbers from list of integers
def even(x):
    if x % 2 == 0:
        return 1
arr = [1,2,3,4,5,6,7]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
# zip(iterable1, iterable2, ..... , iterableN)
# It returns list of tuples, the tuple is generated as each position elements from all iterables.
name = ['ravi','srinu','siva']
age = [20,21,22]
print(list(zip(name,age)))
```

```python
# enumerate(iterable): It returns the list of tuples of index and value
arr = [10,20,30]
print(list(enumerate(arr)))
```

```python
arr = list(map(int,input().split()))
target = int(input())
for ind,val in enumerate(arr):
    if target == val:
        print(ind)
        break
else:
    print(-1)
```

```python
# any(iterable): It returns True when any one iterable element's boolean value is True
arr = [1,2,4,0,0,0]
print(any(arr))
print(any([0,0,0,0,0]))
print(all(arr))
# all(iterable): It returns True when all iterable element's boolean value is True
print(all([1,2,3,4]))
print(all([True,False,True]))
print(all([False,False,False]))
```

```python
# isinstance(obj,class_name) : It checks weather the object is belongs to particular class or not
print(isinstance(10,int))
print(isinstance(12.25,int))
```
