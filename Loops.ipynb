```python
# Even -> square , odd -> 
lst = list(map(int,input().split()))
for num in lst:
    if num % 2 == 0:
        print(num ** 2)
    else:
        print(num + 1)
```

```python
# Fizz(3) Buzz(5) FizzBuzz(3 and 5)
lst = list(map(int,input().split()))
for num in lst:
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

```python
# Count the vowels and consonants in a string.
word = input()
vowel = "aeiouAEIOU"
v,c = 0,0
for ch in word:
    if ch in vowel:
        v += 1
    else:
        c += 1
print(c)
print(v)
        
```

```python
# Count the vowels and consonants in a string excluding digits and special characters.
word = input()
vowel = "aeiouAEIOU"
v,c = 0,0
for ch in word:
    if ch.isalpha():
        if ch in vowel:
            v += 1
        else:
            c += 1
print(c)
print(v)
        
```

```python
n = int(input())
fac = 2
factors = [1,n]
for i in range(2,(n//2) + 1):
    if n % i == 0:
        factors += [i]
        fac += 1
        break
if fac == 2:
    print("Prime")
    print(factors)
else:
    print("Not Prime")
```
