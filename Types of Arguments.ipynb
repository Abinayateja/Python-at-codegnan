```python
# Types of Arguments
# 1. Positional Arguments : Here arguments are assigns to parameters as per the sequential position.
# function definition block
def sub(x:int,y:int)->int:
    z = x - y
    return z

#function calling block
a,b = 20, 10
print(sub(a,b))
print(sub(b,a))
```

```python
# Keyword Arguments : arguments are expliitly passed with parameter names
# function definition block
def sub(x,y:int)->int:
    z = x - y
    return z

#function calling block
a,b = 20, 10
print(sub(x = a,y = b))
print(sub(y = b,x = a))
```

```python
# function definition block
def add(x,y:int)->int:
    z = x + y
    return z

#function calling block
a,b = 20, 10
print(add(x = a,y = b))
print(add(y = b,x = a))
print(add(5,y = 10))
```

```python
# Variable length positional arguments
def add(x,y = 0,*args):
    s = x + y
    for val in args:
        s += val
    return s
print(add(10, 20, 1, 2 , 3, 4,5))
```

```python
# In python every type function calling block shares the arguments refference to parameters
# If the arguments are immutable data types, then behave like pass by value and if arguments
# are mutable data types(i.e list, set, dict) then behave like pass by refference
# Pass by Value
def fun(x):
    print("Address before addition :",id(x))
    x += 2
    print("Address after addition :",id(x))
    print(x)
a = 10
print("Before Fnction Execution 'a' value:",a)
fun(a)
print("After Function Execution 'a' value:",a)
```

```python
# Pass by Refference.
def fun(lst):
    print("Address before updation :",id(lst))
    lst.append('Abinayateja')
    print("Address after updation :",id(lst))
    print(lst)
lst = [10,20,30]
print("Before Function Execution 'a' value:",lst)
fun(lst)
print("After Function Execution 'a' value:",lst)
```

```python
lst = [[1,2,3,4],[5,6,7,8],[2,4,6,8],[1,3,5,7]]
diagonal_sum = 0
for i in range(len(lst)):
    for j in range(len(lst[0])):
        if i == j:
            diagonal_sum += lst[i][j]
print(diagonal_sum)
```

```python
mat = [[1,2,3,4],[5,6,7,8],[2,4,6,8],[1,3,5,7]]
total_sum = 0
for i in range(len(mat)):
    for j in range(len(mat[0])):
        if i == j or (i + j == len(mat) - 1):
            total_sum += mat[i][j]
print(total_sum)
```

```python
# Scopes of Variables
# 1. Local Scope
# 2. Enclosing Scope
# 3. Global Scope
# 4. Builtin Scope

x = 10 # Global Scope variable
def outer():
    y = 20 # Enclosing Scope variable
    def inner():
        z = 30 # Local Scope variable
        print(x,y,z) # print() --> Builtin Scope
    inner()
outer()
```

```python
# Using global Scope variable in local scope
x = 10
def outer():
    y = x + 1
    print("Local Scope variable Value: ", y)
outer()
print("Global Scope variable value: ",x)
```

```python
# Updating the Global Scope variable in Local Scope
# can't update global variable in local scope directly.
x = 10
def outer():
    x = x + 1
    print("Local Scope variable Value: ", x)
outer()
print("Global Scope variable value: ",x)
```

```python
# Updating the Global Scope variable in Local Scope using 'global()'' keyword
x = 100
def outer():
    global x
    x = x + 1
    print("Local Scope variable Value: ", x)
outer()
print("Global Scope variable value: ",x)
```

```python
# Updating the Enclosing scope variable in local scope
x = 10
def outer():
    y = 20
    def inner():
        z = 30
        y = y + 10
        print(y)
    inner()
outer()
```

```python
# Updating the Enclosing scope variable in local scope by using nonlocal
y = 10
def outer():
    y = 20
    def inner():
        nonlocal y
        z = 30
        y = y + 10
        print(y)
    inner()
outer()
print(y)
```
