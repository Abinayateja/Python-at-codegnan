# Tuple

```python
# characteristics
# Ordered Collection
t = (1,2,3,[10,30,40],23)
t[3][2] # Nested Indexing
```

```python
# Immutable --> After performing operation (length won't change) & (remaining elements positions won't change)
t = (1,2,3,[10,20,40],23)
t[3].append(100)
print(t)
```

```python
# Empty Tuple
t1 = ()
t2 = tuple()
print(type(t1),type(t2))
print(t1,t2)
```

```python
# Single Element Representation
t1 = (10)
t2 = (10,)
l1 = [10]
print(type(t1),type(t2),type(l1))
```

```python
# Tuple Operations
# Tuple Concatination
# Tuple Repetition
t = (1,2,3,4,5)
print(t * 3)
print(t + (100,100,100))
```

```python
# Methods
# index()
# count()
print(t2.count(10))
#print(t2.index(2)) # Value Error
print(t2.index(10))
```

```python
# Builtin Functions
t = (1,2,3,4,5)
print(len(t))
print(min(t))
print(max(t))
print(sum(t))
print(sum(t) / len(t)) # Average
```
