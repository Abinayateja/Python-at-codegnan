# Exception Handling

`Exception :` Exception is nothing but it is an error whenever an exception is occurs it stops the program execution and it returns Type of Exception.
and followed by Exception message.

### Types of Exceptions / Errors
1. Compile time Errors (or) Syntax Errors.
2. Runtime Errors
3. Logical Errors

<br>
1. `Compile Time Errors/ Syntax errors :` The syntax errors raises during the compilation time of the program those errors we can call it as Syntax Errors / Compile Time Errors.

Ex: Indentation Errors, Syntax Erros.
<br><br>
2. `Runtime Erros:` The Runtime Errors raises during runtime of a program those errors called as Runtime Erros.
- We cannot handle Syntax Errors but we can handle the Runtime Errors by using Execption handling blocks (i.e, try-except)

Ex: Key Error, Value Error, Type Error, Attribute Error, Index Error, ZeroDivisionError, etc

`Exception Handling Blocks:`<br>
`Syntax:` <br>
   `             try:`<br>
    `                  # Block code that might contain exeption.`<br>
   `             except:`<br>
    `                   # Excception Handling Block.`<br>
    `             else:`<br>
    `                   # This Block executes when there is no exception raised in try Block.`<br>
    `             finally:`<br>
    `                   # This Block always executes.`

```python
# Try Block with Exception.
try:
    lst = [1,2,3]
    print(lst[10])
    print("Hi")
except:
    print("Something  wrong in try Block.")
else:
    print("No Exception raised in try block")
finally:
    print("Program Done.")
    
```

```python
# Try Block without the Exception
try:
    lst = [1,2,3]
    print(lst[2])
    print("Hi")
except:
    print("Something  wrong in try Block.")
else:
    print("No Exception raised in try block")
finally:
    print("Program Done.")
    
```

```python
# We cant Handle Syntax Errors.
try:
    print("Hello')
except:
    print("Something  wrong in try Block.")
else:
    print("No Exception raised in try block")
finally:
    print("Program Done.")
    
```
