# File Handling
`File :` The file is used to store data in electronic format in system

`File Operations :`
1. Opening a file --> open()
2. Reading a file --> read()
3. writing into a file --> write()
4. closing a file --. close()

`Opening a file :` To open a file by using open() function in that we need to pass file name and file opening mode.

`Syntax:` <br>
   `             file_obj = open(file_name,mode)`<br>
    `             # read/write operations`<br>
   `             #close file`<br>

`File opening modes` :
 1. r --> (read) --> Opening file in read mode.(if not exists `FileNotFound`error)
 2. w --> (write) --> Opening a file in write mode. (if not exists in the path it will create a new file in the path otherwise it erases the previous content).
 3. a --> (append) --> Opening a file in append mode. (if not exists in the path it will create a new file in the path otherwise it will add the content at the end of the file to the previous content)
 4. x --> create --> creates new file.
 5. b --> binary --> rb(read binary),wb(write binary) --> This mode used to perform operations on binary files for writing `wb` mode and for reading `rb` mode.
 6. r+ --> read + write
 7. w+ --> write + read
 8. a+ --> append + read


`Reading Content from File` : To perform the read operations on a file for that we have to open the file in `r` mode and then we can perform following operations:
1. read() --> It returns entire file content as single string.
2. readline() --> This method returns first line.
3. readlines() --> This method returns list of lines.


`Writing into file` : If you open a file in write mode we can perform following operations:
1. write(string) : It adds string into a file.
2. writelines(list_of_lines) : It will add list of lines to the file.

`Closing a file` : After performing read/write operations on a file we need to close the file by using close() function, otherwise data loss will happen.


### Note:
- File opening using `with open()`<br>
`Syntax:` <br>
   `             with open(file_name,mode) as file_obj`<br>
    `             # file operations`<br>
