## Encapsulation
- Encapsulation means, wrapping the methods and attributes in single unit cthat is class. The main aim of Encapsulation is to restrict the direct access of attributes and methods.
- This Type of Restriction can implement by using `Access Modifiers`

#### Types of Access Modifiers
1. Public
2. Protected
3. Private

- In python we dont have any keywords to represent it as public,protected and private, but by using underscore(_) can define.
  1. Public : The Public Members (i.e attributes + methods) does not start with underscore. These members can be accessed anywhere in the program.
  2. Protected : The Protected Members (i.e attributes + methods) starts with a single underscore(_). These members can be accessed inside the class and in sub classes also. We can't access them outside the class.
  3. Private : The Private Members (i.e attributes + methods) starts with double underscore(__). These members can be accessed only inside the class. We can't access them in subclasses or outside the class.

- Python does not support high security like java.Here we can access the protected members and private members in outside the class by using following ways.
  - `To access the Protected Member :`<br>
obj._protected-member-name <br><br>
  - `To access the Private Member by using name Mangling :`<br>
obj._Class-Name__private-member-name <br><br>

- The Advantage of Encapsulation is to provide the security for data and functionalities.


```python
# Encapsulation Program (Restricting the Direct access of sensitive information of the bank.)

class Bank:
    def __init__(self,name,account,password,balance):
        self.name = name
        self.account = account
        self.__password = password
        self._balance = balance
    def __total_users(self): # getter method
        return "Total users : 101"
    def get_balance(self):
        return f"Current Balance is:{self._balance}"
    def deposit(self,deposit_amount): # setter method
        self._balance += deposit_amount
        return f"{deposit_amount} Deposited Successfully"

u1 = Bank("Ravi",101,"ravi@123",10000)
print(u1.name)
print(u1.get_balance())
print(u1.deposit(5000))
print(u1.get_balance())
#print(u1.__password) # we cant access thee private members outside the class.
print("password", u1._Bank__password)
print(u1._balance) # Prottected members acess in outside the class
print(u1._Bank__total_users()) # Name Mangling : Syntax: obj._classname__attribute
```
