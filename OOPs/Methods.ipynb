### Methods

`Methods:`
The Function definition is implemented inside the class those functions called as methods

`Types of Methods:`
1. Instance Methods
2. Class Methods
3. Static Methods

`1. Instance Method: `
- The method definition takes first parameter as 'self' thoes methods called as `Instance Method`.
- These method will work only on instance attributes.

  `Syntax:` <br>
   `             class class_name:`<br>
    `                  def method_name(self,parameters):` <br>
    `                      # block code`
  
`2. Class Method:`
The method definition starts with `@classmethod` decorator and it takes first parameter as `cls` those methods called as `Class Methods`. This methods works on class Attributes.

`Syntax:` <br>
   `             class class_name:`<br>
   `                     @classmethod `<br>
    `                    def method_name(cls,parametes):` <br>
    `                      # block code`

`3. Static Method:`
The method definition starts with `@staticmethod` decorator and it doesn't take first parameter as `self or cls` those methods called as `Static methos`. This methods workds on except instance attributes and class attributes.

`Syntax:` <br>
   `             class class_name:`<br>
   `                     @staticmethod `<br>
    `                    def method_name(parameters):` <br>
    `                      # block code`

- `__init__:`
This is also called initialization constructor or magic method which is used to initialize the instance attributes and it takes `self` as the first parameter and It is called whenever an `object` is created for the `class`.
