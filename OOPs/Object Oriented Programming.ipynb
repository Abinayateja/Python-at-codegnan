# Object Oriented Programming
`Procedural`
- Hard to Maintain.
- Low Level Modularity.
- Low Security.
- High Memory.

`OOPs`
- Higher Control over the functionalities.
- Limitted Access and Highly Secure.
- Higher Modularity.
- Low Memory.
# -------------------------------------------------------------
### OOPs Main Components
1. Class
2. Object
3. Encapsulation ---> Security.
4. Inheritance ---> Code Reusability.
5. Polymorphism ---> Different Behaviours.
6. Abstration ---> Hiding Complex Implementation.

# -------------------------------------------------------------

`Class` :
- A Class is a Blue Print / Template of a real world Object.
- Class is a Collection of Attributes(Data) and Methods.

`Syntax` :
            class class_name:
                # attributes
                # methods
# ---------------------------------------------------------------

`Object` :
- Object is an Instance of a Class and it is a Real World Item.

`Syntax` :
            object = class_name(value)
# ---------------------------------------------------------------

`Attributes`:
- The variables are defined inside the class those variables called as Attributes.

`Types of Attributes` :
1. Class Attributes---> The variables are defined outside of initialization constructor. These attributes data is common for every object of the class.
2. Instance Attributes---> The variables are defined inside the initialization constructor/Magic method. These attributes Data different to every obj.

`self` ---> Instance

```python
# define Person Class
class Student:
    school = "XYZ School" # Class Attribute

    def __init__(self,name,age): # Initialization Constructor (Magic Method).
        self.name = name # Instance Attribute
        self.age = age # Instance Attribute
        print("Student Class __init__ method")
    
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old."

s1 = Student("Ravi",20)
s2 = Student("Janu",19)
print(s1.name)
print(s1.age)
print(s2.name)
print(s2.age)
print(s1.school)
print(s2.school)
print(s1.details())
print(s2.details())
```
