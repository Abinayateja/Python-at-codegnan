1. `Method Overriding` :
- A method defined in the child class that is already exists in parent class with the same method name.
- Whenever if you call that method through the child class object the child class method. this is known as MEthod Overriding.

2. `Method Overriding` :
Directly python doesnot support method overloading we can achieve same functionality through default arguments,variable length positional, or variable length keyword arguments.


```python
# Method Overriding :
class Animal:
    def voice(self):
        return "Animals have different sounds"
class Dog(Animal):
    def voice(self):
        return "Dog barks as Bow Bow..."

obj = Dog()
print(obj.voice())
```

```python
# Method Overriding with parameters:

class Person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name}.I am {self.age} years old."

class Student(Person):
    def __init__(self,std_id,name,age,gender,std):
        super().__init__(name,age,gender)
        self.std_id = std_id
        self.std = std
    def details(self,std_id):
        return f"My Student id : {self.std_id} and My name is {self.name}. I am {self.age} years old. and I am studying in {self.std}"

obj = Student(1001,"Abinayateja",21,"Male","B-Tech")
print(obj.details(1001))
# print(obj.details())
        
```

```python
class Math:
    def add(self,a,b):
        return a + b
    def add(self,a,b,c):
        return a + b + c
    def add(self,a,b,c,d):
        return a + b + c + d
obj = Math()
print(obj.add(1,2,3,4))
    
```

```python
## Method Overloading using Default arguments
#directly we cant do method overloading in python but we can achieve the same functionality using  default arguments or Arbitrary arguments.
class Math:
    def add(self,a = 0,b = 0,c = 0,d = 0):
        return a + b + c + d
obj = Math()
print(obj.add(1,2,3,4))
    
```

```python
# Magic methods or Special Methods
class Point:
    def __init__(self,x):
        self.x = x
p1 = Point(5)
print(p1)
```

```python
# __str__() magic method
# if a method contain no print statement then it will see for the str magic method
# if it is there then it will return the output of the str method definition and otherwise it will return the reference of the class object.
class Point:
    def __init__(self,x):
        self.x = x
    def __str__(self):
        return f"this is Point lass and Point value is {self.x}"
p1 = Point(5)
print(p1)
```

```python
# __str__() magic method
# __add__() magic method
class Point:
    def __init__(self,x):
        self.x = x
    def __str__(self):
        return f"this is Point lass and Point value is {self.x}"
    def __add__(self,x):
        return self.x + self.x
p1 = Point(4)
p2 = Point(6)
p3 = p1 + p2
print(p3)
```

```python
# __str__() magic method
# __add__() magic method
class Point:
    def __init__(self,x):
        self.x = x
    def __str__(self):
        return f"this is Point lass and Point value is {self.x}"
    def __add__(self,x):
        return Point(self.x + self.x)
p1 = Point(4)
p2 = Point(6)
p3 = p1 + p2
print(p3)
```

```python
# __str__() magic method
# __add__() magic method
class Point:
    def __init__(self,x):
        self.x = x
    def __str__(self):
        return f"this is Point lass and Point value is {self.x}"
    def __mul__(self,other1):
        return self.x * other1.x
p1 = Point(4)
p2 = Point(6)
p3 = p1 * p2
print(p3)
```

```python
# __str__() magic method
# __truediv__() magic method
class Point:
    def __init__(self,x):
        self.x = x
    def __str__(self):
        return f"this is Point lass and Point value is {self.x}"
    def __truediv__(self,other1):
        return self.x / other1.x
p1 = Point(6)
p2 = Point(2)
p3 = p1 / p2
print(p3)
```
