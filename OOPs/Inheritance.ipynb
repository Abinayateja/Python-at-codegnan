## Inheritance
In Python, a class get properties fro the another classs this process called as Inheritance.

- A child class is inherited from parent class<br>
`Child Class`: A class which acquires the methods and attributes of another class called as Child class.<br>
`Parent Class`: A class from which another class acquires the methods and attributes called as Parent class.

### Types of Inheritances :

1. Single Inheritance : A `child class` get properties from `single parent class `.This is called as Single Inheritance
<br>


`Syntax:` <br>
   `             class parent_class_name:`<br>
    `                        # Class Members`<br>
   `             class child_class_name(parent_class_name):`<br>
    `                        # Class Members`

2. Multi Level Inheritance : A class(i.e `child class`) get properties from another class(i.e ` parent class `) that class get properties from some another class (i.e `grand_parent`)
<br>


`Syntax:` <br>
   `             class parent_class_name:`<br>
    `                        # Class Members`<br>
   `             class child_class_name(parent_class_name):`<br>
    `                        # Class Members`
   `             class grand_child_class_name(child_class_name):`<br>
    `                        # Class Members`  
     `           class gret_grand_child_class_name(grand_child_class_name):`<br>
    `                        # Class Members` 

3. Multiple Inheritance : A `Single child class` get properties from `Multiple parent classes `.This is called as Multiple Inheritance
<br>


`Syntax:` <br>
   `             class parent_1_class_name:`<br>
    `                        # Class Members`<br>
   `             class parent_2_class_name:`<br>
    `                        # Class Members`<br>
    `             class parent_3_class_name():`<br>
    `                        # Class Members`<br>
    
   `             class child_class_name(parent_1_class_name, parent_2_class_name, parent_3_class_name):`<br>
    `                        # Class Members`


4. Hierarchical Inheritance : `Multiple child classes` get properties from `single parent class`.


`Syntax:` <br>
   `             class parent_1_class_name:`<br>
    `                        # Class Members`<br>
   `             class parent_2_class_name(parent_1_class_name):`<br>
    `                        # Class Members`<br>
    `             class parent_3_class_name(parent_1_class_name):`<br>
    `                        # Class Members`<br>
    
   `             class child_class_name(parent_1_class_name, parent_2_class, parent_3_class):`<br>
    `                        # Class Members`

5. Hybrid Inheritance : Combination of `two or more` types of Inheritances.
`Syntax:` <br>
   `             class parent_1_class_name:`<br>
    `                        # Class Members`<br>
   `             class parent_2_class_name:`<br>
    `                        # Class Members`<br>
    `             class parent_3_class_name:`<br>
    `                        # Class Members`<br>
    
   `             class child_class_name(parent_1_class_name, parent_2_class, parent_3_class):`<br>
    `                        # Class Members`

```python
# Types of Inheritances
# Single Inheritance
# Without child class implementation.
# When child class don't have constructor, in that case parent class constructor behave like child class constructor.
class Person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old."
class Student(Person):
    pass

s = Student("Bhanu", 20, "Female")
print(s.name)
print(s.details())
```

```python
# Types of Inheritances
# Single Inheritance
# Without child class implementation.
# When child class don't have constructor, in that case parent class constructor behave like child class constructor.
class Person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old."
class Student(Person):
    def __init__(self,stdid,name,age,gender,branch):
        super().__init__(name,age,gender)
        self.branch = branch
        self.stdid = stdid

s = Student(101, "Bhanu", 20, "Female","ECE")
print(s.name)
print(s.branch)
print(s.details())
```
